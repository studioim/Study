```python
# 7.3.3
import numpy as np
result = []
while True:
    x = np.random.normal(1, 50, 100)
    result.append(x)
    if len(result) > 49:
        break
```

```python
std_mean = []
for x in result:
    std_mean.append(np.mean(x))
std_mean
```

```python
np.std(result[0])
```

```python
sns.distplot(result[0])
plt.xlim(-150, 150)
plt.show()
```
