```python
# PCA(Principal Component Analysis, 주성분 분석)

from sklearn.datasets import load_iris
iris = load_iris() # 아이리스 데이터 뽑기
```

```python
N = 10 #  앞의 10송이만 선택
X = iris.data[:N, :2] # 꽃받침 길이와 꽃받침 폭만 선택
```

```python
X # 꽃받침 길이, 꽃받침 폭 순
```

```python
# 이 데이터를 PCA 돌려보자
from sklearn.decomposition import PCA

pca1 = PCA(n_components=1) # 1차원으로 축소시키는 PCA 객체를 만들어라
```

```python
X_low = pca1.fit_transform(X) # 저차원 데이터, 10개짜리 1차원 데이터
X_low # 잠재변수, 원점을 데이터 가운데 모여있는쪽으로 옮긴 뒤 길이를 재다보니 마이너스 결과값이 많은 것
```

```python
# 잠재변수의 위치가 정확히 어딘지 모르겠다. 이를 다시 역변환해서 확인
X2 = pca1.inverse_transform(X_low)
X2 # 원래 좌표 X와 비슷
```

```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
faces_all = fetch_olivetti_faces()
#%matplotlib inline
```

```python
K = 1 # 20번 인물만 선택
face = faces_all.images[faces_all.target == K]
pca3 = PCA(n_components=2)

N = 2
M = 5
fig = plt.figure(figsize=(10,5))
face_mean = pca3.mean_.reshape(64, 64)
face_p1 = pca3.components_[0].reshape(64, 64)
face_p2 = pca3.components_[1].reshape(64, 64)

plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k - 5) if k < 5 else 1.5 * (k - 4)
        ax.imshow(face_mean + w * face_p1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title('주성분1의 비중={}'.format(w))
    plt.suptitle("평균 얼굴에 주성분1을 더한 사진")
    plt.tight_layout()
    plt.show()
```
