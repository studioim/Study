```python
# 베이즈추정법 활용한 베르누이 분포의 모수 추정
# 연습문제 9.3.1
# 상품 A(좋아요 2, 싫어요 1)와 상품 B(좋아요 60, 싫어요 40)의 베르누이 모수의 사후분포의 하이퍼파라미터를 구하고 각각 사후분포 확률밀도 함수를 그려라.

xx = np.linspace(0, 1, 1000)
a0, b0 = 2 + 1, 1 + 1
plt.plot(xx, sp.stats.beta(a0, b0).pdf(xx), c='r', ls='--', label='상품A의 사후분포')
a1, b1 = 60 + 1, 40 + 1
plt.plot(xx, sp.stats.beta(a1, b1).pdf(xx), c='g', ls='-', label='상품B의 사후분포')
plt.legend()
plt.show
```
