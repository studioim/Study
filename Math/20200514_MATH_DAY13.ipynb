# 데이터 분석의 소개
- 데이터 주어졌을 때 데이터 간의 관계를 파악하거나
- 파악된 관계를 사용해 우리가 원하는 새로운 출력 데이터를 만들어내는 과정
- 분석 목적에 따라 예측, 클러스터링, 모사(approximation) 등

### 예측
- 가장 많이 활용
- 숫자 문서 이미지 음성 영상 등 여러가지 입력 데이터를 넣어서 우리가 원하는 출력을 뽑을 수 있으면 예측이라고 한다
- 부동산 위치 주거환경 건축연도 등 입력하면 해당 부동산의 가치를 추정할 수 있다
- 꽃잎의 길이 너비 외형적 특징 주면 해당하는 식물의 종을 알아낸다
- 얼굴 사진을 주면 해당하는 사람의 이름을 출력한다
- 현재 바둑돌의 위치를 주면 다음 바둑돌의 위치를 예측

### 내가 풀 수 있었으면 하는 문제? 3가지 정도 생각해봐라
- 데이터에는 입력 데이터와 출력 데이터가 있다
- 독립변수, 특징 피처, 설명변수 등등으로 부른다
- 출력데이터는 종속변수라고 부른다. 카테고리 값이면 라벨, 또는 클래스라고도 한다
- ~ 등, 같은 것이라는 용ㅇ어를 사용하지 않고 정확히 입출력 데이터의 개수를 정해야 한다.
- 입력이나 출력 데이터값의 숫자 예를 제시하면 더욱 좋다.

### 규칙기반 방법과 학습기반 방법
- 규칙기반 방법은 규칙을 사람이 만든다. 어떤 입력이 들어오면 어떤 출력이 나오는지를 결정하는 규칙이나 알고리즘을 사람이 미리 만들어 놓는 방법.
- 전문가들이 이러한 내용을 이미 알고 있어서 전문가 방법론이라고도 한다. 엑스퍼트 시스템
- 번역: 언어로된 문장이 입력데이터, 출력 데이터는 다른 언어로된 문장.
    - 과거에는 규칙 기반 시스템으로 하려고 했다. rule book을 전문가들이 만들었다. boy -> 소년, girl -> 소녀로 해석해라. 이렇게 만든 룰 북이 사전.
    - 사전과 문법 통해서 번역했다. 결과값이 좋지 않음
- 규칙기반 방법론의 제일 큰 문제: 규칙이 변한다, 일대일 대응이 아니다, 룰이 완벽하지 않은 경우가 현실에 많다. 룰을 전문가들 사람이 만들고 있기 때문에. 사람이 규칙을 만드면 그 규칙을 일일이 다 써줘야 한다. 모든 경우를 다 처리할 수 없기 떄문에 제대로 처리 안 된다.
- 규칙을 만드는 일을 기계한테 시키자가 바로 학습 기반이다.
- 학습을 컴퓨터가 아니라 기계가 하는 것. 학습하려면 데이터 있어야 한다. 비싼 전문가 안 쓰고 값싼 기계를 쓰는 것.

#### 지도학습
- 학습용 데이터 집합이 필요하다. training data set
- 개 사진이 들어가면 개, 고양이 사진 들어가면 고양이 인식하도록 하고 싶다
- 사람이 개와 고양이 구분할 때는 부모님이 알려줬다. 일일이 하나씩 다 가르쳐 줌
- 이런 학습을 기계한테 시키면 그림의 정답은 개다, 고양이다 하나하나 다 알려줘야 함.
- 케이스바이케이스로 규칙이 아닌 답을 다 알려주는 것
- 훈련용 데이터 집합은 정답이 표시된 문제집과 똑같다. 정답이 들어가야 함
    - 정답을 데이터분석에서는 타깃, 목표값이라고 한다
    - 목푯값 붙여주는 것은 지도 시키는 사람이 해줘야 한다 이를 레이블링이라고 한다
- 학습용 입력 데이터와 학습용 출력데이터를 넣고 지도를 시키면서 동시에 학습용 입력데이터를 예측 함수에 넣고 실제로 출력되는 데이터와 학습용 출력데이터를 비교해봐야 한다
- 그리고 이를 피드백 통해 알려줘야 함
- 실제로 지도학습에서 레이블링 답을 붙여주는 과정이 가장 힘듦
    - sentiment analysis
    - 주식 리포트를 보고 좋은 주식인지 아닌지 판단, 뉴스를 보고 좋은 뉴스인지 안 좋은 뉴스인지 실시간으로 판단.
    - 이걸 하기 위해서 이 시스템을 사람이 학습을 시켜줘야 한다. 데이터가 필요하다. 리포트를 가져와서 당신이 원하는 답을 적어 놓으면 지도학습을 시킬 수 있다.
    - 잘못된 예: 규칙기반 시스템으로 성능이 좋지 않자 학습기반 시스템을 하기 위해 규칙기반 시스템으로 레이블링 시켜서 학습용 데이터 만들어 학습기반 방법을 만든다

#### 이미지넷
- 학습용 데이터의 중요성
- 페이페이 리 교수가 이끄는 인공지능 팀이 만들어 2009년 공개한 학습용 데이터셋을 말함
- 만들기 위해 인터넷 사이트 직접 개설, AWS의 mechanical turk를 활용(단순 노가다 작업을 전세계에 아웃소싱 해주는 서비스). 알바 쓰면 단가 비싸니까. 이를 통해 레이블링 완성
- 이후 2010년 이미지 인식 경연대회 개최. 120만장 학습데이터 사용해 20만장으로 정확도 확인
- 처음에는 표준편차가 엄청 컸지만 급격히 줄어듦. 2년 만인 2012년 인공신경망 이용한 딥러닝 방법 사용해 압도적으로 좋은 결과 내놓음. 이때부터 GPU 쓰기 시작했고, 깊은 딥러닝 할 수 있는 몇가지 해결. 이후 미세하게 줄어들면서 5% 안까지 격차 줄여

### 전처리와 인코딩
- 숫자 문서 등을 숫자로 바꿔줘야 한다. 의미있는 중요한 정보만 가지고 있는 숫자 정보로
- 이미지 2차원 데이터로 사용할 수도 있지만 1차원 데이터로 바꿔준다
- 입력 차원
    - 숫자 벡터로 만들었을 때 숫자의 길이, 몇개의 숫자가 돼 실제로 진짜 예측 문제에 들어가냐
    - 입력 차원 정해지면 기본적으로 바꾸기 힘듦.
    - 64개 숫자 입력 받아서 예측하는 형태로 시스템 구축하면 $8*8$ 아닌 이미지는 $8*8$로 바꿔야 함
    - 이미지는 크기 바꾸면 되지만 문서는 길이가 제각각
    - 그래서 문서 데이터 같은 경우에는 BOW(Bag Of Words) 방법을 사용
        - 어떤 문서 안에 어떤 단어가 있는지 없는지만 가지고 판단
        - 순서는 중요하지 않음
    - 문서가 10000개의 단어가 쓰인 문서다. 단어 하나마다 번호를 붙인다. 그리고 나서 문서 봤을 때 1번 단어, 2번 단어 몇번 나왔는지를 숫자 벡터로
    - 그때 생긴 벡터는 1만개의 길이가 됨.

#### 카테고리 값(범주값)과 실수값
- 고양이, 개 등이 카테고리값이 된다. 10, 20은 실수값
- 하나의 카테고리를 클래스라고 얘기한다. 클래스가 두 종류밖에 없으면 바이너리 클래스, 여러종류는 다중 클래스라고 한다
- 원래 카테고리 값의 의미는 두 개를 비교할 수 없다는 것. 개가 큰지 고양이가 큰지 구분 x
- 크기 비교가 가능한데도 일부러 카테고리 값으로 표현 경우도 있다. 학점 등
- 출력하고자 하는 값이 숫자값이면 회귀분석, 카테고리값이면 분류문제라고 한다
- 머신러닝은 대부분 분류문제를 푸는 것. 회귀문제 푸는 머신러닝 방법론이라는 것도 있을 수 있음
- 분류 문제를 푸는 것은 객관식 4지선다 선택하는 것과 비슷하다
- 회귀분석 문제는 보스턴 주택 가격처럼 숫자를 예측하는 것. 카테고리 예측하면 분류문제

- 지도학습은 스스로 규칙을 만들어낸다. 규칙을 만들어내는 방법론은 모형이나 알고리즘에 따라서 다 달라진다.
- 서포트벡터머신이라는 방법론. 특징은 규칙의 경계선이 직선으로 나오게 된다.
    - 잘못된 분류도 생길 수밖에 없음
    - 직선이 아니라 곡선으로 하게 되면 오류를 줄일 수 있다
    - 과연 이게 잘하는 짓인가? 
    - 우리가 하려는 것은 예측을 하려는 것. 지금의 데이터는 학습용 데이터일 뿐. 새로운 데이터에서 맞아야
    - 학습용 데이터에 있는 이상한 데이터 아웃라이어 데이터에 맞춰서 경계선을 맞추면 안 됨
    - 다른 데이터를 많이 모아서 진짜 현상은 어떤지 비교해보는 게 가장 확실한 방법.
        - 교차 검정(cross validation)

#### 비지도 학습(unsupervised learning): 지도학습 따르지 않는 모든 종류의 머신러닝 방법
- 클러스터링
    - 데이터 비슷한 애들이냐 아니냐 기준으로 묶어주는 것
    - 정답이 필요 없다. 문제지를 나눠주는 데 해답지가 없다.
    - 서로 비교해서 비슷한 문제, 비슷한 그림까지는 알 수 있다. 답이 뭔지는 모르지만
    - X데이터는 주어지지만 Y데이터는 주어지지 않음.
    - 예) Affinity propagation
- imputation(impainting)
    - 데이터 3개 알려주고 한 지운 뒤 두개 보고 하나가 뭔지 맞춰보라는 식
    - 빠진 부분을 메꿔줄 수 있다

### 데이터 분석에서 사용하기 위한 패키지
- 머신러닝 패키지
    - statsmodels 패키지
        - 예제 데이터 셋, 모수 추정 및 검정, 회귀분석, 시계열 분석 등 다양
        - R에 있는 여러 패키지 기능을 다 묶어서 파이썬에서 쓰고 싶다해서 만든 것.
        - 기본적으로 R에 있는 것들 중 웬만한건 다 할 수 있다.
        - 여러 사람들이 오랜 시간 만들고 있는 패지키다보니 복잡
        - api라는 서브 패키지 하나 만들어서 제일 쓰기 쉬운 형태로 함수 모양 바꿔서 정리
            - import statsmodels.api as sm
    - sklearn 패키지
        - 머신러닝 교육을 위한 최고의 파이썬 패키지
        - 데이터 셋, 전처리(스케일링, 누락데이터 처리, 텍스트 토큰화), 지도학습(회귀분석, LDA/QDA, 서포트벡터머신, 퍼셉트론, SGD, KNN, 가우스 프로세스, 나이브 베이즈, 의사결정 나무, 랜덤포레스트, 부스팅)
        - 비지도학습(가우스 혼합모형, 클러스터링, PCA), 성능 최적화(교차검증, 특징선택, 하이퍼파라미터 최적화)
        - 처음부터 교육용 패키지
            - 파이썬 만들어지고 난 뒤 넘파이 싸이파이 패키지에 사람들 열광, 쉽게 공학용 코딩할 수 있어서

            - 대부분 없어지고 남아 있는 몇몇의 패키지
            - 성능에 관계 없이 다 쓸 수 있도록 넣어놓은 것
            - 심지어 인공신경망도 있지만 텐서플로우처럼 성능이 좋지는 않다
    - missingno 패키지
