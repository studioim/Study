```python
# 연습문제 8.5.1 카이분제곱분포
# 중심극한 정리
np.random.seed(0)
rv = sp.stats.norm()
M = 2000

plt.subplot(1, 2, 1)
N = 6
x = rv.rvs((N, M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(0, 20)
plt.ylim(0, 250)
plt.title('제곱합의 분포 (N=6)')
plt.xlabel('표본값')
plt.ylabel('개수')
```
